### Which one should you use ? (pyplot vs matplotlib oo method?)

* When ploting quicklly ,okay to use pyplot method.
* When ploting something advance ,use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method

over_50.plot(kind='scatter',x='age',y='chol',c='target');
plt.show();
```

```python
# OO method[Object orianted method] mixed with pyplot method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax = ax) # ax = mean that we are using the OO method and we are going to plot this data on ax axis as ax = ax.
# ax.set_xlim([45,100]) # will change the xlimit to 45-100 (earliar it was 50-75 from above graph)

plt.show();
```

```python
over_50.target.values
```

```python
# OO method[Object orianted method] from scrach

fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

#Custimise
ax.set(title="Heart Disease and Colestrol Levels", xlabel="Age",ylabel="Cholestroll");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");


# Add a horizotal line
ax.axhline(over_50["chol"].mean(),linestyle='--');


plt.show();
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)


# Subplot 1 (age v/s colostroal)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#Set axis
ax0.set(title="Heart Diasase & Colestrol Level",ylabel="Colestrol"); # xlabel="Age" is removed as in above line we have done sharex = True


#Add a legind
ax0.legend(*scatter.legend_elements(),title="Target");

# Add mean line of colestrol
ax0.axhline(over_50["chol"].mean(),linestyle="--")

# Subplot 1 (age v/s colostroal)


scatter2 = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
# set axis
ax1.set(title="Heart Disase V/s Max heart Rate",xlabel="Age",ylabel="Thalach(Max heart Rate)")

# add legend
ax1.legend(*scatter2.legend_elements(),title="Target")

# add line
ax1.axhline(over_50["thalach"].mean(),linestyle="--")


# Add title to entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");

plt.show();
```

```python
# Custamizing x & y axis limitations

# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)


# Subplot 1 (age v/s colostroal)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")

#Set axis
ax0.set(title="Heart Diasase & Colestrol Level",ylabel="Colestrol"); # xlabel="Age" is removed as in above line we have done sharex = True

ax0.set_xlim([50,80])

#Add a legind
ax0.legend(*scatter.legend_elements(),title="Target");

# Add mean line of colestrol
ax0.axhline(over_50["chol"].mean(),linestyle="--")

# Subplot 1 (age v/s colostroal)


scatter2 = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")
# set axis
ax1.set(title="Heart Disase V/s Max heart Rate",xlabel="Age",ylabel="Thalach(Max heart Rate)")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# add legend
ax1.legend(*scatter2.legend_elements(),title="Target")

# add line
ax1.axhline(over_50["thalach"].mean(),linestyle="--")


# Add title to entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");

plt.show();
```
