```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [3,45,12,35,56]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
```

```python
ax.plot(x,y)
```

```python
ax.set(title="graph title",xlabel="x-axis",ylabel="yaxis")
```

```python
plt.show()
```
