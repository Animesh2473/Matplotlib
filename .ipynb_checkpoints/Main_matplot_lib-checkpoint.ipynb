# Matplotlib example workflow

## Making figures with Numpy array

```python
# Create some data

x = np.linspace(0,10,100) # creat evanly spaced data syantax(start,stop,number of sample)
x[:100]
```

```python
# plot the data

fig,ax = plt.subplots();
ax.plot(x,x**2);
plt.show();
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
plt.show();
```

```python
# Makwe aplot from dictonary
nut_butter_prics = {"Almund butter" : 10,
                     "Peanut butter" : 8,
                     "Cashew butter" : 12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prics.keys(),nut_butter_prics.values()); #(x,y)
ax.set(title="Butter prize comparision",xlabel="Types of butter",ylabel="Butter prices")
plt.show();
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prics.keys()),list(nut_butter_prics.values())); #in `barh` we have to convert values in list
plt.show();
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000) # will creat a sample which will give the std noramal distrebution graph

fig, ax = plt.subplots()
ax.hist(x);
plt.show();
```

### Subplots

```python
# Subplot option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize = (10,5))


# Plot to each dirrent axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prics.keys(),nut_butter_prics.values())
ax4.hist(np.random.random(100))
plt.show();
```

```python
# Subplot option 2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot to each diffrent index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prics.keys(),nut_butter_prics.values());
ax[1,1].hist(np.random.randn(100));
plt.show();
```

### Which one should you use ? (pyplot vs matplotlib oo method?)

* When ploting quicklly ,okay to use pyplot method.
* When ploting something advance ,use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method

over_50.plot(kind='scatter',x='age',y='chol',c='target');
plt.show();
```

```python
# OO method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax = ax) # ax = mean that we are using the OO method and we are going to plot this data on ax axis as ax = ax.
# ax.set_xlim([45,100]) # will change the xlimit to 45-100 (earliar it was 50-75 from above graph)

plt.show();
```
